```python
# 1. 상품명별 고유 구매금액 배열 만들기
price_map = sales_df.groupby('상품명')['구매금액'].unique()

# 2. 배수 관계로 추정되는 상품 제외
def is_not_simple_multiple(prices):
    prices = np.sort(prices)
    if len(prices) < 2:
        return False
    for i in range(len(prices)):
        for j in range(i + 1, len(prices)):
            small = prices[i]
            large = prices[j]
            if small == 0:
                continue
            ratio = large / small
            # 정수배이면 제외
            if ratio.is_integer():
                return False
    return True

# 3. 조건을 만족하는 상품과 가격 목록 저장
valid_products = []
for name, prices in price_map.items():
    if is_not_simple_multiple(prices):
        valid_products.append({'상품명': name, '구매금액목록': sorted(prices)})

# 4. DataFrame으로 정리
valid_df = pd.DataFrame(valid_products)

# 5. 결과 
valid_df  # 상위 10개만 보기
```

```python
valid_with_zero = valid_df[
    valid_df['구매금액목록'].apply(lambda x: 0 in x and len(x) > 1)
]

# 결과 확인
print(valid_with_zero)
```
